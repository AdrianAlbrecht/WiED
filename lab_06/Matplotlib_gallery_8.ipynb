```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.axes_grid1.anchored_artists import AnchoredDirectionArrows
import matplotlib.font_manager as fm

fig, ax = plt.subplots(figsize=(15,10))
ax.imshow(np.random.random((8, 16)),cmap = 'gray')

# Simple example
simple_arrow = AnchoredDirectionArrows(ax.transAxes, 'X', 'Y',
                                       arrow_props={'ec': 'black', 'fc': 'none',
                                                    'alpha': 1,'lw': 2})
ax.add_artist(simple_arrow)
simple_arrow = AnchoredDirectionArrows(ax.transAxes, 'X', 'Y',
                                       arrow_props={'ec': 'none', 'fc': 'w',
                                                    'alpha': 1,'lw': 0.4})
ax.add_artist(simple_arrow)

# High contrast arrow
high_contrast_part_1 = AnchoredDirectionArrows(
                            ax.transAxes,
                            '111', r'11$\overline{2}$',
                            loc='upper right',
                            arrow_props={'ec': 'w', 'fc': 'none', 'alpha': 1,
                                         'lw': 2}
                            )
ax.add_artist(high_contrast_part_1)

high_contrast_part_2 = AnchoredDirectionArrows(
                            ax.transAxes,
                            '111', r'11$\overline{2}$',
                            loc='upper right',
                            arrow_props={'ec': 'none', 'fc': 'k'},
                            text_props={'ec': 'w', 'fc': 'k', 'lw': 0.4}
                            )
ax.add_artist(high_contrast_part_2)

# Rotated arrow
fontprops = fm.FontProperties(family='serif')

roatated_arrow = AnchoredDirectionArrows(
                    ax.transAxes,
                    '50', '140',
                    loc='lower left',
                    arrow_props={'ec': 'black', 'fc': 'none',
                                                    'alpha': 1,'lw': 2},
                    angle=50,
                    fontproperties=fontprops
                    )
ax.add_artist(roatated_arrow)

roatated_arrow = AnchoredDirectionArrows(
                    ax.transAxes,
                    '50', '140',
                    loc='lower left',
                    arrow_props={'ec': 'none', 'fc': 'w',
                                                    'alpha': 1,'lw': 0.4},
                    angle=50,
                    fontproperties=fontprops
                    )
ax.add_artist(roatated_arrow)

# Altering arrow directions
a1 = AnchoredDirectionArrows(
        ax.transAxes, 'X', 'Y', loc='center',
        length=-0.15,
        sep_x=0.03, sep_y=0.03,
        color='g'
    )
ax.add_artist(a1)

a2 = AnchoredDirectionArrows(
        ax.transAxes, 'X', ' Y', loc='center left',
        aspect_ratio=-1,
        sep_x=0.01, sep_y=-0.02,
        color='pink'
        )
ax.add_artist(a2)


a3 = AnchoredDirectionArrows(
        ax.transAxes, 'X', 'Y', loc='center right',
        length=-0.15,
        aspect_ratio=-1,
        sep_y=-0.1, sep_x=0.04,
        color='magenta'
        )
ax.add_artist(a3)
plt.title("Some vecotrs on it", fontsize=20, fontweight = 'bold')
plt.xlabel('Some X values', fontsize=15)
plt.ylabel('Some Y values', fontsize=15)
plt.show()
```
